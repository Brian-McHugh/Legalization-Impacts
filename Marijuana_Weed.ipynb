```python
import pandas as pd
from random import randint
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
csv_path1 = "BRFSS_MARIJUANA_DATA_2017.csv"
```

```python
brfss2017_df = pd.read_csv(csv_path1)
```

```python
brfss2017_df.head()
```

```python
brfss2017_df = brfss2017_df.fillna(0)
brfss2017_df.head()
```

```python
brfss2017_df.dtypes
```

```python
brfss2017_df['MARIJANA_17']=brfss2017_df['MARIJANA_17'].replace(88, 0)
brfss2017_df.head()
```

```python
brfss2017_df['MARIJANA_17']=brfss2017_df['MARIJANA_17'].replace(77, 0)
brfss2017_df.head()
```

```python
brfss2017_df['MARIJANA_17']=brfss2017_df['MARIJANA_17'].replace(99, 0)
brfss2017_df.head()
```

```python
USE_grpstate_df=brfss2017_df.groupby('@_STATE') ['MARIJANA_17'].sum()
USE_grpstate_df.head()
```

```python
USE_grpstate_df=USE_grpstate_df.to_csv("USE_grpstate_df.csv", index=True, header=True)
```

```python
HOWUSE17_grpstate_df=brfss2017_df.groupby('@_STATE') ['SMOKE_17', 'EAT_17', 'DRINK_17', 'VAPE_17','DAB_17', 'OTHER_17','DK_HOWUSE_17','RE_HOWUSE_17'].sum()
HOWUSE17_grpstate_df.head()
```

```python
HOWUSE17_grpstate_df=HOWUSE17_grpstate_df.to_csv("HOWUSE17_grpstate_df.csv", index=True, header=True)
```

```python
RSNUSE17_grpstate_df=brfss2017_df.groupby('@_STATE') ['MED_17', 'REC_17', 'BOTH_17', 'DK_RSN_17','RE_RSN_17'].sum()
RSNUSE17_grpstate_df.head()
```

```python
RSNUSE17_grpstate_df=RSNUSE17_grpstate_df.to_csv("RSNUSE17_grpstate_df.csv", index=True, header=True)
```
